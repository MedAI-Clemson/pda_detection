### Consensus Calculations

```python
def compute_agreement(data: pd.DataFrame, metric, agreement_func: callable, return_data_matrix=False):
    df = data[['global_key', 'created_by', metric]].reset_index()
    df_wide = df.pivot(index='global_key', columns='created_by', values=metric)
    df_wide = df_wide.dropna()

    y1 = df_wide.iloc[:, -2]
    y2 = df_wide.iloc[:, -1]

    agreement = agreement_func(y1, y2)

    if return_data_matrix:
        return agreement, df_wide
    return agreement
```

#### Anatomical view

```python
# Percent agreement
compute_agreement(df_consensus, metric='anatomical_view', agreement_func=lambda x, y: (x == y).mean())
```

```python
# Cohen kappa statistic
from sklearn.metrics import cohen_kappa_score

kappa, df_wide = compute_agreement(df_consensus, metric='anatomical_view', agreement_func=cohen_kappa_score, return_data_matrix=True)
kappa
```

## Make ground truth set

```python
# create a df with the non-consensus labels
consensus_ids = df_consensus.global_key.unique()
df_nonconsensus = df[~df.global_key.isin(consensus_ids)]
df_nonconsensus
```

```python
# de-duplicate the consensus set

# first get the the set with complete agreement
print("Total number of unique consensus labels:", len(df_consensus.global_key.unique()))

# the set of ids with complete agreement on collapsed anatomical view
_, df_wide_anatomical_view = compute_agreement(df_consensus, metric='collapsed_anatomical_view', agreement_func=cohen_kappa_score, return_data_matrix=True)
ids_agree_anatomical_view = df_wide_anatomical_view[df_wide_anatomical_view.iloc[:, -2] == df_wide_anatomical_view.iloc[:, -1]].index
print("Agree on collapsed anatomical view:", len(ids_agree_anatomical_view))

# of these, find those that also agree on modality
_, df_wide_imaging_modality = compute_agreement(df_consensus[df_consensus.global_key.isin(ids_agree_anatomical_view)], metric='collapsed_anatomical_view', agreement_func=cohen_kappa_score, return_data_matrix=True)
ids_agree_imaging_modality = df_wide_imaging_modality[df_wide_imaging_modality.iloc[:, -2] == df_wide_imaging_modality.iloc[:, -1]].index
print("Also agree on imaging modality:", len(ids_agree_imaging_modality))

# the set of ids with complete agreement is the intersection of these two sets
ids_agree = set(ids_agree_imaging_modality)

# the set of ids with some disagreement
ids_disagree = set(df_consensus.global_key.unique()) - ids_agree

# the part of the consensus set with complete agreement
df_consensus_agree = df_consensus[df_consensus.global_key.isin(ids_agree)]

# the part of the consensus set with some disagreement
df_consensus_disagree = df_consensus[df_consensus.global_key.isin(ids_disagree)]
```

```python
# change pandas setting to display all rows
pd.set_option('display.max_rows', None)
df_consensus_agree.sort_values(['global_key', 'created_by'])
# check that all the rows do, in fact, represent agreement
```

```python
df_consensus_disagree.sort_values(['global_key', 'created_by'])
# check that all the rows do, in fact, represent disagreement
```

```python
assert set(df_consensus.global_key.unique()) == set(df_consensus_agree.global_key.unique()) | set(df_consensus_disagree.global_key.unique()), "Lost some ids"
```

```python
# de-duplicate the agreed set
df_consensus_agree_mode = df_consensus_agree.drop_duplicates(subset=['global_key'])
```

```python
# combine the non-consensus and agreed sets
df_ground_truth = pd.concat([df_consensus_agree_mode, df_nonconsensus]).sort_values('global_key').reset_index(drop=True)
```
