```python
import glob
import os
from dotenv import load_dotenv
from labelbox import Client

load_dotenv()

source_dir = "/project/dane2/wficai/pda/external_validation/Boston/exports/"

mp4_files = glob.glob(source_dir + "/**/*.mp4", recursive=True)

root_url = "https://fast-videos.s3.us-east-2.amazonaws.com/pda/external_validation/boston/"
```

```python
mp4_files
```

```python
def create_lbx_dict(file_path):
    obj_key = file_path.split(source_dir)[-1]
    row_data = f"{root_url}{obj_key}"
    global_id = file_path.split('/')[-1].split('.mp4')[0]

    return {
        "row_data": row_data,
        "global_key": global_id,
        "media_type": "VIDEO"
    }
```

```python
assets = [create_lbx_dict(file) for file in mp4_files]
assets[:5]
```

```python
client = Client(api_key=os.getenv("LBX_API_KEY"))
dataset = client.create_dataset(name="Boston External Validation")
```

```python
task = dataset.create_data_rows(assets)
task.wait_till_done()
print(task.errors)
```
