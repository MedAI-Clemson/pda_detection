```python
def extract_video_metadata(file_path):
    cv2video = cv2.VideoCapture(file_path)
    height = cv2video.get(cv2.CAP_PROP_FRAME_HEIGHT)
    width  = cv2video.get(cv2.CAP_PROP_FRAME_WIDTH)
    
    framecount = cv2video.get(cv2.CAP_PROP_FRAME_COUNT ) 
    fps = cv2video.get(cv2.CAP_PROP_FPS)
    
    return height, width, framecount, fps
```

# Function to make pda look like echonet

```python
import pandas as pd
import cv2
import numpy as np
import os

def extract_video_metadata_df(row):
    file_path = row["FileName"]
    cv2video = cv2.VideoCapture(file_path)
    # print(f"Starting video {file_path}")
    row["FrameHeight"] = cv2video.get(cv2.CAP_PROP_FRAME_HEIGHT)
    row["FrameWidth"]  = cv2video.get(cv2.CAP_PROP_FRAME_WIDTH)
    
    row["NumberOfFrames"] = cv2video.get(cv2.CAP_PROP_FRAME_COUNT)
    row["FPS"] = cv2video.get(cv2.CAP_PROP_FPS)
    
    return row
```

```python
def pda_to_echonet(csv_file):
    # Read CSV file
    df_pda = pd.read_csv(csv_file)
    
    # Drop views we don't want
    df_pda = df_pda[(df_pda.view != "nonPDAView") & (df_pda["mode"] != "2d")]
    
    # Create a new DF to store the conents we want
    pda_columns = ["FileName", "PatientID", "Label", "ExternalID", "Mode", "FrameHeight","FrameWidth","FPS","NumberOfFrames", "Split"]
    new_df = pd.DataFrame(data=None, columns=pda_columns)
    new_df["FileName"] = df_pda["mp4_path"]
    new_df["Label"] = df_pda["patient_type"]
    new_df["PatientID"] = df_pda["patient_id"]
    new_df["ExternalID"] = df_pda["external_id"]
    new_df["Mode"] = df_pda["mode"]
    new_df["Split"] = df_pda["Split"]
    new_df["patient_sample_5"] = df_pda["patient_sample_5"]
    new_df["patient_sample_12"] = df_pda["patient_sample_12"]
    new_df["patient_sample_20"] = df_pda["patient_sample_20"]
    new_df["patient_sample_28"] = df_pda["patient_sample_28"]
    new_df["patient_sample_36"] = df_pda["patient_sample_36"]
    new_df["patient_sample_44"] = df_pda["patient_sample_44"]
    
    
    # Drop Duplicates
    new_df = new_df.drop_duplicates("FileName")
    
    # Run file information get
    df_final = new_df.apply(extract_video_metadata_df, axis=1)
    
    # Split into different distributions
    distrib = ["patient_sample_5","patient_sample_12","patient_sample_20","patient_sample_28", "patient_sample_36", "patient_sample_44"]
    
    for split in distrib:
        print("#####################")
        print(f"Performing split {split}")
        path_to_output = f"./output_{split}/"
        if not os.path.exists(path_to_output):
            os.mkdir(path_to_output)
        df_sample = df_final[((df_final.Split == 'TRAIN') & (df_final[split])) |\
                             df_final.Split.isin({'VAL', 'TEST'})]
        df_sample.to_csv(f"{path_to_output}{split}_file.csv")
        print(df_sample.Split.value_counts())
    
    
    
```
