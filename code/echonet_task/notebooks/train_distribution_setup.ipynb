# Put it in a function instead

```python
def create_sub_samples(df_train, df_val, df_test, percent, output_path="/zfs/wficai/Data/echonet_data/"):
    print(f"Create split: {percent}%")
    # Sample the df_train for split
    df_train_sub = resample(df_train, replace=False, n_samples = int(len(df_train)*percent), random_state=42)
    # Sample the df_val for split
    df_val_sub = resample(df_val, replace=False, n_samples = int(len(df_val)*percent), random_state=42)
    # Append all dfs
    df_out = df_train_sub.append(df_val_sub)
    df_out = df_out.append(df_test)
    # Save to disk
    df_out.to_csv(f"{output_path}train_val_test_{percent}.csv")
    print(f"train samples: {len(df_train_sub)}")
    print(f"val samples: {len(df_val_sub)}")
    print(len(df_out))
    
```

# Create samples with all validation data

```python
import pandas as pd
from sklearn.utils import resample
```

```python
def create_sub_samples_full_val(df_train, df_val, df_test, percent, output_path="/zfs/wficai/Data/echonet_data/"):
    print(f"Create split: {percent}")
    # Sample the df_train for split
    df_train_sub = resample(df_train, replace=False, n_samples = int(len(df_train)*percent), random_state=42)
    # Sample the df_val for split
    df_val_sub = resample(df_val, replace=False, n_samples = int(len(df_val)), random_state=42)
    # Append all dfs
    df_out = df_train_sub.append(df_val_sub)
    df_out = df_out.append(df_test)
    # Save to disk
    df_out.to_csv(f"{output_path}train_fullval_test_{percent}.csv")
    print(f"train samples: {len(df_train_sub)}")
    print(f"val samples: {len(df_val_sub)}")
    print(f"test sampels: {len(df_test)}")
    print(len(df_out))
```

```python
df_train = pd.read_csv("/zfs/wficai/Data/echonet_data/TRAIN.csv")
df_val = pd.read_csv("/zfs/wficai/Data/echonet_data/VAL.csv")
df_test = pd.read_csv("/zfs/wficai/Data/echonet_data/TEST.csv")
percents = [0.01,0.05,0.1,0.25,0.5,0.75]
```

```python
for percent in percents:
    create_sub_samples_full_val(df_train, df_val, df_test, percent)
```
